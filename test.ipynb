```python
def compute_nb_parameters(hidden_size, size_in = 16, size_out = 3):

    return (size_in+1) * hidden_size + (hidden_size+1)*hidden_size*2 + (hidden_size+1) * size_out

compute_nb_parameters(44), compute_nb_parameters(66), compute_nb_parameters(88)
```

```python
def reduced_params(in_size, out_size, sigma_sizes):
    total_network = in_size * out_size
    total_reduced_network = in_size * sigma_sizes + out_size*sigma_sizes + sigma_sizes

    print("Le nombre de paramètre pour le réseau entier est : ", total_network)
    print("\nLe nombre de paramètre pour le réseau réduit est : ", total_reduced_network)
    print("\nOn gagne donc ", 100 * (total_network - total_reduced_network) / total_network, "% de nombre de paramètres"  )


def number_params(in_sizes, out_sizes, sigma_sizes):

    total_network = 0
    total_reduced_network = 0
    for in_size, out_size, sigma_sizes in zip(in_sizes, out_sizes, sigma_sizes):
        total_network += in_size * out_size
        total_reduced_network += in_size * sigma_sizes + out_size*sigma_sizes + sigma_sizes

    print("Le nombre de paramètre pour le réseau entier est : ", total_network)
    print("\nLe nombre de paramètre pour le réseau réduit est : ", total_reduced_network)
    print("\nOn gagne donc ", 100 * (total_network - total_reduced_network) / total_network, "% de nombre de paramètres"  )
```

```python
reduced_params(256, 120, 40)
reduced_params(120, 84, 24)
reduced_params(84, 10, 4)
```
